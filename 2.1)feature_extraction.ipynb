```python
#set project path from config.yaml
import pathlib
import yaml

with open("config.yaml", "r") as f:
   config = yaml.safe_load(f)

project_root = pathlib.Path(config["project"]["root_path"])

print("Current project path: ", project_root)
```

```python
dset_wav_path = project_root / "audio_wav_fulldset"
```

```python
import os
import random
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

wav_files = [f for f in os.listdir(dset_wav_path) if f.endswith('.wav')]

#pick a random file
random_wav = random.choice(wav_files)
wav_path = os.path.join(dset_wav_path, random_wav)

y, sr = librosa.load(wav_path, sr=None)  

# Plot waveform (time domain)
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
librosa.display.waveshow(y, sr=sr)
plt.title(f"Waveform of {random_wav}")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

# Plot spectrogram (frequency domain)
plt.subplot(1, 2, 2)
D = np.abs(librosa.stft(y))  # Short-time Fourier transform
librosa.display.specshow(librosa.amplitude_to_db(D, ref=np.max), sr=sr, x_axis='time', y_axis='log')
plt.title("Spectrogram (dB)")
plt.colorbar(format="%+2.0f dB")
plt.tight_layout()
plt.show()
```
